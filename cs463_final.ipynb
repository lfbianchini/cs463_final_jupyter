In this step, we keep only the columns needed for analysis (year, host country, ISO code, and displacement categories). Then we compute a Total_Displacement column by summing all population-of-concern groups. Finally, we clean the ISO codes by uppercasing and trimming whitespace to ensure consistent merging later.

```python
cols_to_keep = [
    "Year",
    "Country of Asylum",
    "Country of Asylum ISO",
    "Refugees",
    "Asylum-seekers",
    "IDPs",
    "Other people in need of international protection",
    "Stateless persons",
    "Host community",
    "Others of concern"
]

refugee_df = refugee_df[cols_to_keep]

displacement_cols = [
    "Refugees",
    "Asylum-seekers",
    "IDPs",
    "Other people in need of international protection",
    "Stateless persons",
    "Host community",
    "Others of concern"
]

refugee_df["Total_Displacement"] = refugee_df[displacement_cols].sum(axis=1)
refugee_df["Country of Asylum ISO"] = refugee_df["Country of Asylum ISO"].str.upper().str.strip()

refugee_df.head()
```

```python
df = disaster_df.copy()


# 2. Clean date columns and parse event years

# Ensure the date column is parsed correctly
df["Date of Event (start)"] = pd.to_datetime(df["Date of Event (start)"], errors="coerce")

# If Year column is missing or inconsistent, extract from date
df["Year"] = df["Year"].fillna(df["Date of Event (start)"].dt.year)


# 3. Standardize ISO codes and country names

df["ISO3"] = df["ISO3"].astype(str).str.upper().str.strip()
df["Country / Territory"] = df["Country / Territory"].astype(str).str.strip()


# 4. Aggregate event-level data into COUNTRY-YEAR totals


# Replace missing displacement values with 0
df["Disaster Internal Displacements"] = df["Disaster Internal Displacements"].fillna(0)

# ---- A. Total disaster displacements per country-year
total_displacements = df.groupby(["ISO3", "Country / Territory", "Year"])["Disaster Internal Displacements"].sum()

# ---- B. Number of disaster events per country-year
event_counts = df.groupby(["ISO3", "Country / Territory", "Year"])["Event ID"].count()

# ---- C. Maximum displacement caused by a single event (per year)
max_event_displacement = df.groupby(["ISO3", "Country / Territory", "Year"])["Disaster Internal Displacements"].max()

# ---- D. Hazard type counts (one-hot encoding  then sum)
hazard_type_counts = pd.get_dummies(df["Hazard Type"])
hazard_type_counts["ISO3"] = df["ISO3"]
hazard_type_counts["Country / Territory"] = df["Country / Territory"]
hazard_type_counts["Year"] = df["Year"]

hazard_type_yearly = hazard_type_counts.groupby(["ISO3", "Country / Territory", "Year"]).sum()


# 5. Merge all aggregated components into one final dataset


# Convert Series to DataFrames for merging
total_displacements = total_displacements.reset_index(name="Total_Disaster_Displacements")
event_counts = event_counts.reset_index(name="Num_Disaster_Events")
max_event_displacement = max_event_displacement.reset_index(name="Max_Event_Displacement")

# Start with total displacements
final_disaster = total_displacements.merge(
    event_counts, on=["ISO3", "Country / Territory", "Year"], how="left"
).merge(
    max_event_displacement, on=["ISO3", "Country / Territory", "Year"], how="left"
).merge(
    hazard_type_yearly.reset_index(), on=["ISO3", "Country / Territory", "Year"], how="left"
)


# 6. Inspect processed disaster dataset

final_disaster.head()
```

Here we load the global CRU climate dataset and link every climate grid cell to its nearest country using a spatial join. We then compute countryyear climate features (mean temperature, precipitation totals, variability, etc.) by aggregating grid values for each country. Finally, we calculate anomalies by subtracting each countrys long-term mean, producing a clean panel dataset of climate indicators over time.

```python
from shapely.geometry import Point

# 1. Load climate .nc dataset
ds = xr.open_dataset(project_folder + "CRU_all_combined.nc")

# Available variables:
# ds["tas"], ds["tasmin"], ds["tasmax"], ds["pr"]
# dims: (time, lat, lon)

# 2. Load downloaded Natural Earth Admin 0 Countries

world = gpd.read_file("ne_countries/ne_110m_admin_0_countries.shp")

# Standardize columns
world = world.rename(columns={
    "ISO_A3": "ISO3",
    "NAME": "Country"
})[["ISO3", "Country", "geometry"]]

world["ISO3"] = world["ISO3"].str.upper().str.strip()


# 3. Create lat/lon point grid for climate dataset
lats = ds["lat"].values
lons = ds["lon"].values

lon_grid, lat_grid = np.meshgrid(lons, lats)

# Flatten grid into a GeoDataFrame of points
points = [Point(lon_grid[i, j], lat_grid[i, j])
          for i in range(len(lats)) for j in range(len(lons))]

grid_gdf = gpd.GeoDataFrame({"geometry": points}, crs="EPSG:4326")

# 4. Spatial join: assign each grid cell to nearest country

# 4A. Reproject both layers to a projected CRS (meters)
world_proj = world.to_crs("EPSG:3857")
grid_proj = grid_gdf.to_crs("EPSG:3857")

# 4B. Nearest-neighbor spatial join
grid_joined = gpd.sjoin_nearest(
    grid_proj,
    world_proj,
    how="left",
    distance_col="dist"
)

# 4C. Drop duplicates that sjoin_nearest sometimes creates
grid_joined = grid_joined.drop_duplicates(subset=["geometry"])

# 4D. Reproject back to WGS84
grid_joined = grid_joined.to_crs("EPSG:4326")

# 4E. Extract ISO codes back into the correct grid order
# IMPORTANT: grid_gdf.index preserves original order
grid_joined = grid_joined.sort_index()

iso_codes = grid_joined["ISO3"].values.reshape(len(lats), len(lons))

# Replace NaNs with placeholder
iso_codes = np.where(pd.isna(iso_codes), "NONE", iso_codes)

# 5. Compute countryyear climate features
records = []

for t in range(len(ds["time"])):
    year = pd.to_datetime(ds["time"].values[t]).year

    # Extract grids for this time slice
    tas = ds["tas"][t].values
    tasmin = ds["tasmin"][t].values
    tasmax = ds["tasmax"][t].values
    pr = ds["pr"][t].values

    # Compute features for each country
    for iso in world["ISO3"].unique():
        mask = (iso_codes == iso)

        if mask.sum() == 0:
            continue

        records.append({
            "ISO3": iso,
            "Year": year,
            "tas_mean": np.nanmean(tas[mask]),
            "tasmin_mean": np.nanmean(tasmin[mask]),
            "tasmax_mean": np.nanmean(tasmax[mask]),
            "pr_sum": np.nansum(pr[mask]),          # total precipitation
            "pr_mean": np.nanmean(pr[mask]),        # NEW: average precipitation
            "tas_std": np.nanstd(tas[mask]),        # temperature variability
            "pr_std": np.nanstd(pr[mask])           # precipitation variability
        })


# Convert to DataFrame
climate_country_year = pd.DataFrame(records)


# 6. Compute climate anomalies (value - long-term mean)

climate_country_year["tas_anomaly"] = climate_country_year.groupby("ISO3")["tas_mean"].transform(
    lambda x: x - x.mean()
)

climate_country_year["pr_anomaly"] = climate_country_year.groupby("ISO3")["pr_sum"].transform(
    lambda x: x - x.mean()
)


# 7. Final result

climate_country_year.head()
```

After computing monthly climate features for each country, we group them by ISO3 and Year to create a single annual record per country. We aggregate temperature fields using the mean, precipitation using yearly sums, and variability measures using averaged monthly variability. After aggregation, we recompute anomalies so they reflect deviations from each countrys long-term yearly averages.

```python
# Aggregate to yearly country climate
climate_country_year = (
    climate_country_year
    .groupby(["ISO3", "Year"], as_index=False)
    .agg({
        "tas_mean": "mean",
        "tasmin_mean": "mean",
        "tasmax_mean": "mean",
        "pr_sum": "sum",     # sum over all 12 months
        "pr_mean": "mean",
        "tas_std": "mean",
        "pr_std": "mean"
    })
)

print("Unique climate rows after aggregation:", len(climate_country_year))

# Recompute anomalies AFTER aggregation
climate_country_year["tas_anomaly"] = climate_country_year.groupby("ISO3")["tas_mean"].transform(lambda x: x - x.mean())
climate_country_year["pr_anomaly"] = climate_country_year.groupby("ISO3")["pr_sum"].transform(lambda x: x - x.mean())
```

Here we prepare each dataset separately (UNHCR displacement, disaster counts, and climate features) by standardizing ISO3 codes and grouping to yearly country-level summaries. After cleaning, we merge all three sources on ISO3 and Year to form one unified panel dataset. Finally, we remove rows lacking displacement targets and save the fully merged dataset for modeling.

```python
# 1. PREPARE UNHCR DISPLACEMENT DATA

unhcr = refugee_df.copy()

# Standardize ISO column to "ISO3"
unhcr = unhcr.rename(columns={
    "Country of Asylum ISO": "ISO3",
    "Country of Asylum": "Country"
})

# Keep required columns only
unhcr = unhcr[["ISO3", "Country", "Year", "Total_Displacement"]]

unhcr_agg = (
    unhcr.groupby(["ISO3", "Year"], as_index=False)
         .agg({
             "Total_Displacement": "sum",
             "Country": "first"    # keep a representative country name
         })
)

print("UNHCR aggregated rows:", len(unhcr_agg))



# 2. PREPARE DISASTER DATA

disaster = final_disaster.copy()

# Ensure ISO3 is string
disaster["ISO3"] = disaster["ISO3"].astype(str)

print("Disaster rows:", len(disaster))



# 3. PREPARE CLIMATE DATA

climate = climate_country_year.copy()

# Ensure ISO3 is string
climate["ISO3"] = climate["ISO3"].astype(str)

print("Climate rows:", len(climate))



# 4. MERGE ALL THREE DATASETS

merged = (
    unhcr_agg
    .merge(disaster, on=["ISO3", "Year"], how="left")
    .merge(climate, on=["ISO3", "Year"], how="left")
)

print("Merged rows (before cleaning):", len(merged))



# 5. CLEAN MISSING TARGET VALUES

merged_clean = merged.dropna(subset=["Total_Displacement"])

print("Merged rows (after removing missing targets):", len(merged_clean))



# 6. OPTIONAL: Clean rows missing climate or disaster information

# merged_clean = merged_clean.dropna()



# 7. CHECK TARGET DISTRIBUTION

print("\nTarget (Total_Displacement) summary:")
print(merged_clean["Total_Displacement"].describe())

merged_clean["Total_Displacement"].hist(bins=50)



# 8. SAVE FINAL MERGED DATASET

merged_clean.to_csv("final_merged_dataset.csv", index=False)

merged_clean.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans


# 1. Select numeric features and standardize


# OPTIONAL: keep an ID column (e.g., country name) if present
id_cols = []
for col in ['country', 'Country', 'iso3', 'ISO3']:
    if col in merged_clean.columns:
        id_cols.append(col)

# Use only numeric feature columns
feature_cols = merged_clean.select_dtypes(include='number').columns.tolist()
X = merged_clean[feature_cols].dropna()

# Save index mapping (in case dropna removed rows)
valid_idx = X.index

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# 2. PCA - reduce to 2 components (for visualization)


pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Explained variance (optional quick check)
print("Explained variance ratio (PC1, PC2):", pca.explained_variance_ratio_)


# 3. K-Means clustering on PCA-reduced data
#    ( can change k if desired)


k = 5  # choose number of clusters
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_pca)


# 4. Attach PCA components and cluster labels back to dataframe


# Create a result DataFrame aligned with rows used in X
results = merged_clean.loc[valid_idx].copy()
results['PC1'] = X_pca[:, 0]
results['PC2'] = X_pca[:, 1]
results['cluster'] = clusters

# Log-transform PCA values to reduce outlier influence
results['PC1_log'] = np.sign(results['PC1']) * np.log1p(abs(results['PC1']))
results['PC2_log'] = np.sign(results['PC2']) * np.log1p(abs(results['PC2']))

print(results[id_cols + ['PC1', 'PC2', 'cluster']][:10] if id_cols else
      results[['PC1', 'PC2', 'cluster']][:10])


# 5. Visualize clusters in PCA space


plt.figure(figsize=(8, 6))
scatter = plt.scatter(results['PC1_log'], results['PC2_log'],
            c=results['cluster'], cmap='viridis', alpha=0.8)
plt.xlabel("log-scaled PC1")
plt.ylabel("log-scaled PC2")
plt.title("Clusters of Countries Based on Climate, Disasters, and Displacement")

if 'country' in results.columns:
    for _, row in results.iterrows():
        plt.text(row['PC1'], row['PC2'], str(row['country']),
                 fontsize=6, alpha=0.7)

plt.colorbar(scatter, label="Cluster")
plt.tight_layout()
plt.show()
```

Here we isolate the climate variables and standardize them before running PCA on the full set. We inspect the cumulative variance to determine how many components capture at least 90% of the climate variability, then rerun PCA with that number of components. These principal components are added back to the dataframe as compact climate descriptors for downstream modeling.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA


# 1. Select climate features

climate_features = [
    "tas_mean",
    "tasmin_mean",
    "tasmax_mean",
    "pr_sum",
    "pr_mean",
    "tas_std",
    "pr_std",
    "tas_anomaly",
    "pr_anomaly"
]

# Keep only rows with all climate data
climate_df = model_df.dropna(subset=climate_features)


# 2. Standardize climate data

scaler = StandardScaler()
X_climate_scaled = scaler.fit_transform(climate_df[climate_features])


# 3. PCA (initial run)

pca = PCA(n_components=None)
climate_pca = pca.fit_transform(X_climate_scaled)

explained = np.cumsum(pca.explained_variance_ratio_)
print("Cumulative explained variance:", explained)


# 4. Choose number of components explaining  90%

k = np.argmax(explained >= 0.90) + 1
print("Number of components to keep:", k)


# 5. PCA with k components

pca_final = PCA(n_components=k)
climate_pca_k = pca_final.fit_transform(X_climate_scaled)


# 6. Add PCs back into dataframe

for i in range(k):
    climate_df[f"Climate_PC{i+1}"] = climate_pca_k[:, i]


# 7. drop original features

# climate_df = climate_df.drop(columns=climate_features)

# climate_df is now  new modeling dataframe
climate_df.head()
```

We merge World Bank population data into the climatedisaster dataset and forward/backward fill gaps within each country. Disaster-related fields are cleaned by replacing missing values with zeros, and non-feature columns are removed. The displacement target is log-transformed for stability, and we build the feature matrix by excluding identifiers and target columns. Finally, we apply a time-based train/test split, training on data up to 2015 and testing on 2016 onward.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score


# 0. Copy dataset

df = climate_df.copy()


# 1. Load population data and merge

pop_df = pd.read_csv("population.csv")  # columns: ISO3, Year, Population

# Ensure numeric year
pop_df["Year"] = pd.to_numeric(pop_df["Year"], errors="coerce")

# Merge population into climate dataset
df = df.merge(pop_df, on=["ISO3", "Year"], how="left")

# Fill missing population (per-country)
df["Population"] = df.groupby("ISO3")["Population"].transform(
    lambda s: s.fillna(method="ffill").fillna(method="bfill")
)

# If any weird ones still missing, just drop
df = df[df["Population"].notna()]


# 2. Fill disaster columns with zeros

disaster_cols = [
    "Total_Disaster_Displacements",
    "Num_Disaster_Events",
    "Max_Event_Displacement",
    "Drought",
    "Earthquake",
    "Storm",
    "Flood",
    "Volcanic activity",
    "Landslide",
    "Wildfire"
]

for c in disaster_cols:
    if c in df.columns:
        df[c] = df[c].fillna(0)


# 3. Drop non-feature columns

target = "Total_Displacement"
drop_cols = ["Country", "Country / Territory"]
df = df.drop(columns=drop_cols, errors="ignore")


# 4. Log-transform target

df["log_target"] = np.log1p(df[target])


# 5. Define feature matrix
# (Population is now included automatically)

feature_cols = [
    c for c in df.columns
    if c not in ["ISO3", "Year", target, "log_target"]
]

X = df[feature_cols]
y = df["log_target"]

print("Number of features:", len(feature_cols))
print("Features:", feature_cols)


# 6. Time-based train/test split

train_mask = df["Year"] <= 2015
test_mask  = df["Year"] >= 2016

X_train = X[train_mask]
X_test  = X[test_mask]
y_train = y[train_mask]
y_test  = y[test_mask]

print("Train size:", len(X_train))
print("Test size:", len(X_test))
print("Train years:", df.loc[train_mask, "Year"].min(), "", df.loc[train_mask, "Year"].max())
print("Test years:", df.loc[test_mask, "Year"].min(), "", df.loc[test_mask, "Year"].max())
```

We fit a simple Decision Tree Regressor on the training set and generate predictions in log-space, then convert both predictions and true values back to original displacement units. This allows us to evaluate RMSE and R in real-world terms, while also comparing performance in log-space. The CART model serves as a baseline before moving to more complex ensemble methods.

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

cart = DecisionTreeRegressor(random_state=42)
cart.fit(X_train, y_train)

# 1. Predict in log space
preds_log = cart.predict(X_test)

# 2. Convert predictions + true labels back to original displacement units
preds_raw = np.expm1(preds_log)
y_test_raw = np.expm1(y_test)

# 3. Compute RMSE and R in original units
rmse_raw = np.sqrt(mean_squared_error(y_test_raw, preds_raw))
r2_raw = r2_score(y_test_raw, preds_raw)

print("CART Performance (original units):")
print("  RMSE:", rmse_raw)
print("  R:", r2_raw)

# Optional: also show log-space metrics for comparison
rmse_log = np.sqrt(mean_squared_error(y_test, preds_log))
r2_log = r2_score(y_test, preds_log)

print("\nCART Performance (log space):")
print("  RMSE:", rmse_log)
print("  R:", r2_log)
```

We run a grid search over tree depth and minimum split size to find the best-performing CART configuration. Using 3-fold cross-validation and RMSE as the scoring metric, GridSearchCV selects the optimal combination of parameters, giving us an improved decision tree model for further evaluation

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    "max_depth": [3, 5, 7, 10, 15, None],
    "min_samples_split": [2, 5, 10, 20]
}

grid = GridSearchCV(
    DecisionTreeRegressor(random_state=42),
    param_grid,
    scoring="neg_root_mean_squared_error",
    cv=3,
    n_jobs=-1
)

grid.fit(X_train, y_train)

best_cart = grid.best_estimator_

print("Best Parameters:", grid.best_params_)
```

Using the best hyperparameters from GridSearchCV, we generate new predictions and convert them back to original displacement units. We then compute RMSE and R to measure real-world performance, and also show the log-space metrics for comparison. This lets us directly assess the improvement gained from tuning the CART model.

```python
tuned_preds_log = best_cart.predict(X_test)

# Convert log  original units
tuned_preds_raw = np.expm1(tuned_preds_log)
y_test_raw = np.expm1(y_test)

# RMSE in raw units (manual sqrt)
rmse_raw = np.sqrt(mean_squared_error(y_test_raw, tuned_preds_raw))
r2_raw = r2_score(y_test_raw, tuned_preds_raw)

print("Tuned CART Performance (original units):")
print("  RMSE:", rmse_raw)
print("  R:", r2_raw)

# Optional: log-space metrics
rmse_log = np.sqrt(mean_squared_error(y_test, tuned_preds_log))
r2_log = r2_score(y_test, tuned_preds_log)

print("\nTuned CART Performance (log space):")
print("  RMSE:", rmse_log)
print("  R:", r2_log)
```

# **Random Forest Regression**

We rebuild the full modeling dataset by merging population data, filling missing values, and setting all disaster-related fields to zero where needed. After log-transforming displacement, we construct the feature matrix (now including Population) and apply a temporal train/test split. A tuned Random Forest is trained and evaluated in both raw and log units, and we extract feature importances to see which variables most strongly influence displacement predictions.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import pandas as pd


# 0. Copy dataset

df = climate_df.copy()


# 1. Merge population

pop_df = pd.read_csv("population.csv")  # ISO3, Year, Population
pop_df["Year"] = pd.to_numeric(pop_df["Year"], errors="coerce")

df = df.merge(pop_df, on=["ISO3", "Year"], how="left")

# Fill missing population values per country (forward/backward fill)
df["Population"] = df.groupby("ISO3")["Population"].transform(
    lambda s: s.fillna(method="ffill").fillna(method="bfill")
)

# Drop leftover missing
df = df[df["Population"].notna()]


# 2. Fill disaster columns with zeros

disaster_cols = [
    "Total_Disaster_Displacements",
    "Num_Disaster_Events",
    "Max_Event_Displacement",
    "Drought",
    "Earthquake",
    "Storm",
    "Flood",
    "Volcanic activity",
    "Landslide",
    "Wildfire"
]

for c in disaster_cols:
    if c in df.columns:
        df[c] = df[c].fillna(0)


# 3. Drop non-feature columns

target = "Total_Displacement"
drop_cols = ["Country", "Country / Territory"]
df = df.drop(columns=drop_cols, errors="ignore")


# 4. Log-transform target

df["log_target"] = np.log1p(df[target])


# 5. Define feature matrix (Population now included)

feature_cols = [
    c for c in df.columns
    if c not in ["ISO3", "Year", target, "log_target"]
]

print("FEATURE COLS NOW INCLUDE POPULATION?")
print("Population" in feature_cols)

X = df[feature_cols]
y = df["log_target"]


# 6. Time split

train_mask = df["Year"] <= 2015
test_mask  = df["Year"] >= 2016

X_train = X[train_mask]
X_test  = X[test_mask]
y_train = y[train_mask]
y_test  = y[test_mask]


# 7. Train model

rf = RandomForestRegressor(
    n_estimators=600,
    min_samples_split=4,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1,
    oob_score=True,
)

rf.fit(X_train, y_train)


# 8. Evaluate

y_pred_log = rf.predict(X_test)
y_pred = np.expm1(y_pred_log)

y_test_raw = np.expm1(y_test)

rmse = np.sqrt(mean_squared_error(y_test_raw, y_pred))
r2 = r2_score(y_test_raw, y_pred)

r2_log = r2_score(y_test, y_pred_log)
print("Log-space R:", r2_log)

rmse_log = np.sqrt(mean_squared_error(y_test, y_pred_log))
print("Log-space RMSE:", rmse_log)



print("RMSE:", rmse)
print("R:", r2)


# 9. Feature importance

importances = pd.DataFrame({
    "Feature": feature_cols,
    "Importance": rf.feature_importances_
}).sort_values(by="Importance", ascending=False)

importances.head(20)
```
